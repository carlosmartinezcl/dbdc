 * `density_food_type`: Representa el cuociente entre le total de restaurants por tipo de comida y ciudad, respecto del total de restaurantes de la ciudad. (`n_rest/total_rest`)
 * `ratio_review`: Representa el cociente entre a valoración del restaurant por tipo de comida y ciudad, respecto de la valoración promedio de los resturants de la misma ciudad. (`review_prom/review_prom_city`)

`````````
#P4a) (4pts) Genere una tabla llamada resumen, que contenga la siguiente información
# city: Ciudad
# food_type: Tipo de comida
# n_rest: Cantidad de restaurantes por cada ciudad y tipo de comida.
# review_prom: Valoración promedio por cada ciudad y tipo de comida.
# total_rest: Total de restaurantes por cada ciudad (se puede repetir el valor por cada tipo de comida).
# review_prom_city: Valoración promedio de los restaurantes por cada ciudad (se puede repetir el valor por cada tipo de comida).

# n_rest
my_nrest <- tbl_aux %>%
  group_by(city, food_type) %>%
  summarise(n_rest = n())
`````````

`````````
# review_prom
my_review_prom <- tbl_aux %>%
  group_by(city, food_type) %>%
  summarise(review_prom = mean(review, na.rm = TRUE))
`````````

`````````
# total_rest
my_total_rest <- tbl_aux %>%
  group_by(city) %>%
  summarise(total_rest = n_distinct(id_rest))
`````````

`````````
# review_prom_city
my_review_prom_city <- tbl_aux %>%
  group_by(city) %>%
  summarise( review_prom_city = mean(review, na.rm = TRUE))
`````````

`````````
# Generando tabla
tbl_4a <-  left_join(my_nrest, my_review_prom, by = c("city", "food_type")) %>%
  left_join(my_total_rest, by = "city") %>%
  left_join(my_review_prom_city, by = "city")
`````````

`````````
# Para que quede igual a la tabla ejemplo
tbl_4a <- tbl_4a %>% arrange(desc(n_rest))
head(tbl_4a)
`````````

`````````
#P4b) (2pts) Basado en la tabla anterior, construya dos nuevas columnas llamadas density_food_type y ratio_review que contengan la siguiente información:
# density_food_type: Representa el cuociente entre le total de restaurants por tipo de comida y ciudad, respecto del total de restaurantes de la ciudad. (n_rest/total_rest)
# ratio_review: Representa el cociente entre a valoración del restaurant por tipo de comida y ciudad, respecto de la valoración promedio de los resturants de la misma ciudad. (review_prom/review_prom_city)

tbl_4a <- tbl_4a %>%
   mutate(density_food_type = n_rest / total_rest)
tbl_4a <- tbl_4a %>%
   mutate(ratio_review = review_prom / review_prom_city)
head(tbl_4a)
`````````

`````````
# P4c) (3pts) Mediante un gráfico de dispersión, muestre la relación entre density_food_type y ratio_review.
# Investigue sobre el parámetro alpha dentro de la capa geométrica para una mejor visualizaciónd de los puntos.
# Adicionalmente añada una curva de tendencia y, con base en él, indique si cabe la posibilidad de establecer
# algún tipo de dependencia entre density_food_type y ratio_review.

my_grafico <- ggplot(tbl_4a, aes(x = density_food_type, y = ratio_review)) +
  geom_point(alpha = 0.6, color = "red", size = 2) +
  geom_smooth(method = "lm", se = FALSE) +
  labs(title = "Relación Density Food Type y Ratio Review",
       x = "Density Food Typew",
       y = "Ratio Review") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 18),
    axis.title = element_text(size = 12),
    axis.text = element_text(size = 10)
  )

print(my_grafico)
`````````

`````````
# P5a) (3pts) En la tabla resumen creada en P4a), genere una nueva columna llamada type_review, que contenga
# "review alto", si ratio_review >= 1 y "review bajo" ratio_review < 1. ¿Qué indica esta variable? Comente.
tbl_4a <- tbl_4a %>%
  mutate(type_review = ifelse(ratio_review >= 1, "review alto", "review bajo"))
head(tbl_4a)
`````````

¿Qué indica esta variable? Comente.<br>
R: La columna `type_review` indica que una  ciudad puede tener una calificación alta o baja, basado en si está por sobre o por debajo del promedio de las calificaciones del tipo de comida de esa ciudad.<br>
Por ejemplo:<br>

*   San Francisco tiene una baja calificación en comida china.
*   San Francisco tiene una alta calificación en comida italiana.

A las perdonas que visiten San Francisco se les puede recomendar visitar restaurants de comida italiana.

`````````
# P5b) (3pts) Para cada type_review, muestre a través de un gráfico de cajas (boxplot),
# la distribución de la densidad del tipo de comidadensity_food_type.

qrtl <- tbl_4a %>%
  group_by(type_review) %>%
  summarise(Q1 = quantile(density_food_type, 0.25),
            Q2 = quantile(density_food_type, 0.50),
            Q3 = quantile(density_food_type, 0.75),
            Q4 = quantile(density_food_type, 1))
qrtl
# Convertir en data.frame
qrtl_df <- as.data.frame(qrtl)

# Crear el gráfico de cajas
grafico_caja <- ggplot(tbl_4a, aes(x = type_review, y = density_food_type)) +
  geom_boxplot(fill = "#02FAD5", color = "#8902FA") +
  labs(title = "Distribución de densidad del tipo de comida por tipo de revisión",
       x = "Tipo de Revisión",
       y = "Densidad del Tipo de Comida") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 18),
    axis.title = element_text(size = 12),
    axis.text = element_text(size = 10)
  )

# Mostrar el gráfico de cajas
print(grafico_caja)
`````````

Hay una mayor dispersión en los review altos.<br>
